```python
import numpy as np
import pandas as pd
import os

file_list = []
for dirname, _, filenames in os.walk('../input/'):
    for filename in filenames:
        file_list.append(os.path.join(dirname, filename))
        print(os.path.join(dirname, filename))
        
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import gc
warnings.simplefilter('ignore')
my_pal = sns.color_palette(n_colors=10)
matplotlib.rcParams['figure.dpi'] = 100
sns.set()
%matplotlib inline
```

```python
input_path = '../input/'

specs = pd.read_csv(os.path.join(input_path, 'specs.csv'))
train = pd.read_csv(os.path.join(input_path, 'train.csv'))
train_labels = pd.read_csv(os.path.join(input_path, 'train_labels.csv'))
test = pd.read_csv(os.path.join(input_path, 'test.csv'))

specs = reduce_mem_usage(specs)
train = reduce_mem_usage(train)
train_labels = reduce_mem_usage(train_labels)
test = reduce_mem_usage(test)
```

```python
print(
    'specs : {}'.format(specs.shape),
    'train : {}'.format(train.shape),
    'train_labels : {}'.format(train_labels.shape),
    'test : {}'.format(test.shape), 
     sep='\n')
```

```python
train_ = train.sample(1000000)
test_ = test.sample(1000000)
```

```python
train_ = train.merge(specs, on='event_id', how='left')
test_ = test.merge(specs, on='event_id', how='left')

train_data = (train_.count() / len(train_)).sort_values().values
ind = np.arange(len(train_data))
width = 0.35

fig, axes = plt.subplots(1,1,figsize=(14, 6))
tr = axes.bar(ind, train_data, width, color='royalblue')

test_data = (test_.count() / len(test_)).sort_values().values
tt = axes.bar(ind+width, test_data, width, color='seagreen')

axes.set_ylabel('Amount of data available');
axes.set_xticks(ind + width / 2)
axes.set_xticklabels((train_.count() / len(train_)).sort_values().index, rotation=90)
axes.set_title('Amount of available data and NaNs', fontsize=16)
axes.legend([tr, tt], ['Train', 'Test'])
plt.show()
```

```python
train_labels.groupby('accuracy_group')['game_session'].count().plot(kind='barh', figsize=(15, 5), title='Target (accuracy group)')
plt.show()
```

```python
sns.pairplot(train_labels, hue='accuracy_group')
plt.show()
```

```python
train['event_id_as_int'] = train['event_id'].apply(lambda x: int(x, 16))
train['game_session_as_int'] = train['game_session'].apply(lambda x: int(x, 16))
```

```python
# Format and make date / hour features
train['timestamp'] = pd.to_datetime(train['timestamp'])
train['date'] = train['timestamp'].dt.date
train['hour'] = train['timestamp'].dt.hour
train['weekday_name'] = train['timestamp'].dt.weekday_name
# Same for test
test['timestamp'] = pd.to_datetime(test['timestamp'])
test['date'] = test['timestamp'].dt.date
test['hour'] = test['timestamp'].dt.hour
test['weekday_name'] = test['timestamp'].dt.weekday_name
```

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 9), dpi=100)
tmp_D = train.groupby('date')['event_id'].count()
tmp_D.plot(ax=axes[0], label='By date', alpha=0.8, color=my_pal[2])
axes[0].set_title('Numer of Event Observations by Day of Week', fontsize=16)

tmp_H = train.groupby('hour')['event_id'].count()
tmp_H.plot(ax=axes[1], label='By date', alpha=0.8, color=my_pal[1])
axes[1].set_title('Numer of Event Observations by hour', fontsize=16)

cols_week = ['Monday','Tuesday','Wednesday', 'Thursday','Friday','Saturday', 'Sunday']
tmp_W = train.groupby('weekday_name')['event_id'].count().T[cols_week]
tmp_W.plot(ax=axes[2], label='By date', alpha=0.8, color=my_pal[3])
axes[2].set_title('Numer of Event Observations by weekday', fontsize=16)

plt.subplots_adjust(hspace=0.45)
plt.show()
```

```python
train.groupby('installation_id') \
    .count()['event_id'] \
    .plot(kind='hist',
          bins=40,
          color=my_pal[4],
          figsize=(15, 5),
         title='Count of Observations by installation_id')
plt.show()
```

```python
train.groupby('installation_id') \
    .count()['event_id'] \
    .apply(np.log1p) \
    .plot(kind='hist',
          bins=40,
          color=my_pal[6],
         figsize=(15, 5),
         title='Log(Count) of Observations by installation_id')
plt.show()
```

```python
train.groupby('installation_id') \
    .count()['event_id'].sort_values(ascending=False).head(5)
```

```python
train.query('installation_id == "f1c21eda"') \
    .set_index('timestamp')['event_code'] \
    .plot(figsize=(15, 5),
          title='installation_id #f1c21eda event Id - event code vs time',
         style='.',
         color=my_pal[8])
plt.show()
```

```python
train.groupby('event_code') \
    .count()['event_id'] \
    .sort_values() \
    .plot(kind='bar',
         figsize=(15, 5),
         title='Count of different event codes.')
plt.show()
```

```python
train['game_time'].apply(np.log1p) \
    .plot(kind='hist',
          figsize=(15, 5),
          bins=100,
          title='Log Transform of game_time',
          color=my_pal[1])
plt.show()
```

```python
train.groupby('title')['event_id'] \
    .count() \
    .sort_values() \
    .plot(kind='barh',
          title='Count of Observation by Game/Video title',
         figsize=(15, 15))
plt.show()
```

```python
train.groupby('type')['event_id'] \
    .count() \
    .sort_values() \
    .plot(kind='bar',
          figsize=(15, 4),
          title='Count by Type',
          color=my_pal[2])
plt.show()
```

```python
train.groupby('world')['event_id'] \
    .count() \
    .sort_values() \
    .plot(kind='bar',
          figsize=(15, 4),
          title='Count by World',
          color=my_pal[3])
plt.show()
```

```python
train['log1p_game_time'] = train['game_time'].apply(np.log1p)

fig, ax = plt.subplots(figsize=(15, 5))
sns.catplot(x="type", y="log1p_game_time",
            data=train.sample(10000), alpha=0.5, ax=ax);
ax.set_title('Distribution of log1p(game_time) by Type')
plt.close()
plt.show()
fig, ax = plt.subplots(figsize=(15, 5))
sns.catplot(x="world", y="log1p_game_time",
            data=train.sample(10000), alpha=0.5, ax=ax);
ax.set_title('Distribution of log1p(game_time) by World')
plt.close()
plt.show()
```
